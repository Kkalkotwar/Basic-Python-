#  Handling Missing Values Pandas interpolate 2

```python
import pandas as pd
```

```python
df=pd.read_csv('students_10.csv')
df
```

```python
df.interpolate() # bydefault it will fill missing vaues linearly
```

```python
df.interpolate(limit=1)# bydefault it will fill only 1 missing vaues linearly 
```

```python
df.interpolate(limit=2)# bydefault it will fill only 2 missing vaues linearly
```

```python
df.interpolate(limit_direction='backward')
```

```python
df.interpolate(limit=1,limit_direction='backward')# bydefault it will fill only 1 missing vaues
```

```python
df.interpolate(limit=2,limit_direction='backward')# bydefault it will fill only 2 missing vaues
```

```python
df.interpolate(limit=1,limit_direction='both')# bydefault it will fill only 1 missing vaues by both backward and forward method
```

```python
df.interpolate(limit_area='inside')
```

```python
df.interpolate(limit_area='outside')
```

```python
df.interpolate(inplace=True) # bydeafult it will fill linearly and inplace =True will update the dataframe
df
```
