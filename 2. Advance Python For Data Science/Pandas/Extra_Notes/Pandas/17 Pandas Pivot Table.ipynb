# Pandas Pivot Table

```python
import pandas as pd
```

```python
df=pd.read_csv('Movies_ratings.csv')
df
```

```python
df.pivot_table(index = 'Year')
```

```python
df.pivot_table(index = 'Genre')
```

```python
df.pivot_table(index = 'Genre',columns='Year')
```

```python
df.pivot_table(index = 'Genre',columns='Year',aggfunc='count')
```

```python
df.pivot_table(index = 'Genre',columns='Year',aggfunc='max')
```

```python
df.pivot_table(index = 'Genre',columns='Year',aggfunc='sum')
```

```python
df.pivot_table(index = 'Genre',columns='Year',fill_value='none')
```

```python
df.pivot_table(index = 'Genre',columns='Year',fill_value=0)
```

```python
df.pivot_table(index = 'Genre',columns='Year',fill_value=0,margins=True)
```

```python
df.pivot_table(index = 'Genre',columns='Year',aggfunc='sum',fill_value=0,margins=True)
```

```python
df.pivot_table(index = 'Year',columns='Budget(M$)')
```

```python
df.pivot_table(index = 'Year',columns='Budget(M$)',aggfunc='count')
```

```python
df.pivot_table(index = 'Year',columns='Budget(M$)',aggfunc='max')
```

```python
df.pivot_table(index = 'Year',columns='Budget(M$)',aggfunc='sum')
```

```python
df.pivot_table(index = 'Year',columns='Budget(M$)',fill_value='none')
```

```python
df.pivot_table(index = 'Year',columns='Budget(M$)',fill_value=0)
```

```python
df.pivot_table(index = 'Year',columns='Budget(M$)',fill_value=0,margins=True)
```

```python
df.pivot_table(index = 'Year',columns='Budget(M$)',aggfunc='sum',fill_value=0,margins=True)
```
