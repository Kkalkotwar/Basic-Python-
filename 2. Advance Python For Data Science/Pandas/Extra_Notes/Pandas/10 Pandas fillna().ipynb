# Pandas fillna() Method

```python
import pandas as pd
```

```python
df=pd.read_csv('Fortune_100.csv')
df
```

```python
df.fillna(0)
```

```python
df.fillna(2)
```

```python
df.fillna({'Industry':'none','Growth':0})
```

```python
df.fillna(method='ffill')
```

```python
df.fillna(method='pad')
```

```python
df.fillna(method='bfill')
```

```python
df.fillna(axis=0)# it will give error  bcoz axis=0 means row , and we have to specify a value to be fill 
```

```python
df.fillna(axis=0,method='ffill')
```

```python
df.fillna(axis=1,method='ffill')
```

```python
df.fillna(axis=0,method='bfill')
```

```python
df.fillna(axis=1,method='bfill')
```

```python
df.fillna(0,limit=1) # it will fill only one nan values with 0 ineach column where nan value is present
```

```python
df.fillna(0,limit=2) # it will fill only two nan values with 0 ineach column where nan value is present
```

```python
df.fillna(method='ffill',limit=1) # it will fill only one nan values with 0 ineach column where nan value is present
```

```python
df.fillna(5,inplace= True) # inplace= True means it will update df dataframe with 5 in place of nan values
df
```
