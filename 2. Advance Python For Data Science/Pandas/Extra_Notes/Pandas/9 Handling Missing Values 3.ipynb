```python
import pandas as pd
```

```python
df=pd.read_csv('Fortune_100.csv')
df
```

```python
df.shape
```

```python
df.dropna()
```

```python
df.dropna(axis=1)# drop all colums having NaN Values
```

```python
df.dropna(how='any')# bydefault removr all rows having NaN values
```

```python
df.dropna(axis=1,how='any')
```

```python
df.dropna(how='all')# it remove all rows and columns of dataset if all values are NaN
```

```python
df1=pd.read_csv('Fortune_1000.csv')
df1
```

```python
df1.dropna(how='all')# it remove all rows and columns of dataset if all values are NaN
```

```python
df
```

```python
df.dropna(thresh=1) # print all rows if minimum one non null value is present in that row
```

```python
df.dropna(thresh=2) # print all rows if minimum 2 non null value is present in that row
```

```python
df.dropna(thresh=4) # print all rows if minimum 4 non null value is present in that row
```

```python
df.dropna(thresh=7)# print all rows if minimum 7 non null value is present in that row
```

```python
df.dropna(subset=['ID']) # drop all rows of columns ID if having NaN Values
```

```python
df.dropna(subset=['Industry'])  # drop all rows of columns Industry if having NaN Values
```

```python
df.dropna(subset=['Growth']) # drop all rows of columns Growth if having NaN Values
```

```python
df.dropna(inplace=True) # update the dataframe df if inplace=True
df
```
