# Pandas interpolate

```python
import pandas as pd
```

```python
df=pd.read_csv('students_10.csv')
df
```

```python
df.interpolate() # it will fill only numerical value
```

```python
df.interpolate(method='linear')
```

```python
df.interpolate(method='time')
```

```python
type(df.Date[0])
```

```python
df=pd.read_csv('students_10.csv', parse_dates=['Date'])
df
```

```python
df=pd.read_csv('students_10.csv', parse_dates=['Date'],index_col='Date')
df
```

```python
df.interpolate(method='time')
```

```python
df.interpolate(method='index')
```

```python
df=pd.read_csv('students_10.csv', parse_dates=['Date'],index_col='Class')
df
```

```python
df.interpolate(method='index')
```

```python
df.interpolate(method='nearest')
```

```python
df=pd.read_csv('students_10.csv', parse_dates=['Date'],index_col='Date')
df
```

```python
df.interpolate(method='polynomial',order=2)
```
